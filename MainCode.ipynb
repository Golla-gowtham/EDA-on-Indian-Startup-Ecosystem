```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("startup_funding.csv", encoding='utf-8')

# Rename columns for easier use
df.rename(columns={
    'Startup Name': 'StartupName',
    'Industry Vertical': 'IndustryVertical',
    'SubVertical': 'SubVertical',
    'City  Location': 'CityLocation',
    'Investors Name': 'InvestorsName',
    'InvestmentnType': 'InvestmentType',
    'Amount in USD': 'AmountInUSD',
    'Date dd/mm/yyyy': 'Date'
}, inplace=True)

# Strip whitespace and standardize text columns
df['StartupName'] = df['StartupName'].str.strip()
df['CityLocation'] = df['CityLocation'].str.title().str.strip()
df['IndustryVertical'] = df['IndustryVertical'].str.strip()
df['SubVertical'] = df['SubVertical'].str.strip()
df['InvestorsName'] = df['InvestorsName'].str.strip()

# Clean AmountInUSD column
df['AmountInUSD'] = df['AmountInUSD'].str.replace(",", "").replace("Undisclosed", np.nan)
df['AmountInUSD'] = pd.to_numeric(df['AmountInUSD'], errors='coerce')

# Convert Date column
df['Date'] = pd.to_datetime(df['Date'], errors='coerce', dayfirst=True)
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

# Drop rows with missing Amount or Date
df.dropna(subset=['AmountInUSD', 'Date'], inplace=True)

print("Cleaned Data Shape:", df.shape)
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Year', palette='Set2')
plt.title("Number of Startups Funded Each Year")
plt.xlabel("Year")
plt.ylabel("Startup Count")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
yearly_funding = df.groupby('Year')['AmountInUSD'].sum().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=yearly_funding, x='Year', y='AmountInUSD', palette='viridis')
plt.title("Total Funding Amount per Year (USD)")
plt.ylabel("Amount in USD (Billions)")
plt.xlabel("Year")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
top_cities = df['CityLocation'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_cities.index, y=top_cities.values, palette='magma')
plt.title("Top 10 Cities by Number of Funded Startups")
plt.xlabel("City")
plt.ylabel("Number of Startups")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
industry_counts = df['IndustryVertical'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=industry_counts.index, y=industry_counts.values, palette='coolwarm')
plt.title("Top 10 Industries by Number of Startups Funded")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_funded = df.groupby('StartupName')['AmountInUSD'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_funded.values, y=top_funded.index, palette='cubehelix')
plt.title("Top 10 Most Funded Startups")
plt.xlabel("Total Funding (USD)")
plt.tight_layout()
plt.show()
```

```python
corr_df = df[['AmountInUSD', 'Month', 'Year']]
plt.figure(figsize=(8,6))
sns.heatmap(corr_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
